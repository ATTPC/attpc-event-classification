```python
import matplotlib.pyplot as plt
import sklearn.metrics

import ast
```

```python
results_path = '../keras-results/'
```

```python
def getBestCM(best_epoch, contents):
    start_i = 12 + 2*best_epoch - 1
    cm = contents[start_i:start_i + 2]
    return cm

def getBestCMMultic(best_epoch, contents):
    start_i = 12 + 3*best_epoch - 2
    cm = contents[start_i:start_i + 3]
    return cm
```

```python
def binF1(TP, FN, FP, TN):
    F1P = (2*TP)/(2*TP + FN + FP)
    F1N = (2*TN)/(2*TN + FN + FP)
    return [F1P, F1N]

def multiF1(zz, zo, zt, oz, oo, ot, tz, to, tt):
    F1P = (2*zz)/(2*zz + oz + tz + zo + zt)
    F1C = (2*oo)/(2*oo + zo + to + oz + ot)
    F1O = (2*tt)/(2*tt + zt + ot + tz + to)
    return [F1P, F1C, F1O]

def binAcc(TP, FN, FP, TN):
    return (TP + TN)/(TP + FN + FP + TN)

def multiAcc(zz, zo, zt, oz, oo, ot, tz, to, tt):
    return (zz + oo + tt)/(zz + zo + zt + oz + oo + ot + tz + to + tt)
```
