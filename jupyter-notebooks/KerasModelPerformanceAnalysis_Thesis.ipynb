# Sim2Sim

```python
basicNN_sim2sim_pC = open(results_path + 'NN/sim2sim/pC.txt','r')
basicNN_sim2sim_pC_contents = basicNN_sim2sim_pC.read().split('\n')

basicNN_sim2sim_pC_acc = ast.literal_eval(basicNN_sim2sim_pC_contents[1])
basicNN_sim2sim_pC_val_acc = ast.literal_eval(basicNN_sim2sim_pC_contents[4])
basicNN_sim2sim_pC_loss = ast.literal_eval(basicNN_sim2sim_pC_contents[7])
basicNN_sim2sim_pC_val_loss = ast.literal_eval(basicNN_sim2sim_pC_contents[10])
print(basicNN_sim2sim_pC_contents)
```

```python
plt.figure(1)
plt.plot(basicNN_sim2sim_pC_acc)
plt.plot(basicNN_sim2sim_pC_val_acc)
plt.title('Basic NN - Sim2Sim - pC - Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
```

```python
plt.figure(1)
plt.plot(basicNN_sim2sim_pC_loss)
plt.plot(basicNN_sim2sim_pC_val_loss)
plt.title('Basic NN - Sim2Sim - pC - Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
```

```python
best_epoch = 1
start_i = 12 + 2*best_epoch - 1

basicNN_sim2sim_pC_cm = basicNN_sim2sim_pC_contents[start_i:start_i + 2]
print(basicNN_sim2sim_pC_cm)
```
