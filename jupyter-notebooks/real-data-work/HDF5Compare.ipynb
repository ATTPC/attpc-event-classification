```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import scipy as sp
import h5py

import pytpc

from mpl_toolkits.mplot3d import Axes3D
```

```python
data_path = '../../data/real/'
```

```python
filename_peaks = '/reduced_data/run_0130_peaks.h5'
peaks = h5py.File(data_path + filename_peaks, 'r')
```

```python
filename_full = 'run_0130_full.h5'
full = pytpc.HDFDataFile(data_path + filename_full, 'r')
```

```python
filename_less = 'run_0130_less.h5'
less = pytpc.HDFDataFile(data_path + filename_less, 'r')
```

```python
i = 0
```

```python
i = 236

dataset_name = '/reduced_get_events'
data_peaks = peaks[dataset_name]
peaks_evt = data_peaks[str(i)]

evt_full = full[i]
xyzs_full = evt_full.xyzs(peaks_only=True, return_pads=True, baseline_correction=True, cg_times=True)

evt_less = less[i]
xyzs_less = evt_less.xyzs(peaks_only=True, return_pads=True, baseline_correction=True, cg_times=True)
```

```python
fig = plt.figure(figsize=(20, 14))

ax = fig.add_subplot(2, 3, 1, projection='3d')
ax.scatter(peaks_evt[:,0], peaks_evt[:,1], peaks_evt[:,2])
ax.axis('equal')
ax.set_title("Peaks Only")
ax.set_xlim(-275, 275)
ax.set_ylim(-275, 275)
ax.set_zlim(0, 512)

#######################################################################################################################
ax2 = fig.add_subplot(2, 3, 2, projection='3d')
ax2.scatter(xyzs_full[:,0], xyzs_full[:,1], xyzs_full[:,2])
ax2.axis('equal')
ax2.set_title('Full')
ax2.set_xlim(-275, 275)
ax2.set_ylim(-275, 275)
ax2.set_zlim(0, 512)

#######################################################################################################################
ax3 = fig.add_subplot(2, 3, 3, projection='3d')
ax3.scatter(xyzs_less[:,0], xyzs_less[:,1], xyzs_less[:,2])
ax3.axis('equal')
ax3.set_title('Less')
ax3.set_xlim(-275, 275)
ax3.set_ylim(-275, 275)
ax3.set_zlim(0, 512)
```
