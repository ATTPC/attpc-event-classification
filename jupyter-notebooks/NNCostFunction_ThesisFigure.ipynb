```python
import matplotlib.pyplot as plt
import sklearn.metrics
import numpy as np

import ast
```

```python
results_path = '../keras-results/'
```

```python
CNN_real2real_pC = open(results_path + 'CNN/real2real/pC.txt','r')
CNN_real2real_pC_contents = CNN_real2real_pC.read().split('\n')

CNN_real2real_pC_acc = ast.literal_eval(CNN_real2real_pC_contents[1])
CNN_real2real_pC_val_acc = ast.literal_eval(CNN_real2real_pC_contents[4])
CNN_real2real_pC_loss = ast.literal_eval(CNN_real2real_pC_contents[7])
CNN_real2real_pC_val_loss = ast.literal_eval(CNN_real2real_pC_contents[10])
print(CNN_real2real_pC_val_loss)
```

```python
plt.rcParams["axes.labelsize"] = 17
plt.rcParams["axes.titlesize"] = 17
plt.rcParams["xtick.labelsize"] = 17
plt.rcParams["ytick.labelsize"] = 17
plt.rcParams["legend.fontsize"] = 17


best_epoch = 6

fig, ax = plt.subplots(2, 2, squeeze=True, figsize=(22,16))
ax[0][0].plot(CNN_real2real_pC_acc)
ax[0][0].plot(CNN_real2real_pC_val_acc)
ax[0][0].set_title('CNN Accuracy - Tests with Real Data - Proton vs. Carbon')
ax[0][0].set_ylabel('Accuracy')
ax[0][0].set_xlabel('Epoch')
ax[0][0].legend(['Training Set', 'Test Set'], loc='lower right')
ax[0][0].set_ylim(-0.05, 1.05)
ax[0][0].plot(np.full((10,), best_epoch), np.linspace(-0.5, 1.5, 10), c='black')
ax[0][0].set_xticks([best_epoch, 20, 40, 60, 80, 100])
ax[0][0].set_xticklabels(['Optimal', '20', '40', '60', '80', '100'])

ax[0][1].plot(CNN_real2real_pC_loss)
ax[0][1].plot(CNN_real2real_pC_val_loss)
ax[0][1].set_title('CNN Cost - Tests with Real Data - Proton vs. Carbon')
ax[0][1].set_ylabel('Cost')
ax[0][1].set_xlabel('Epoch')
ax[0][1].legend(['Training Set', 'Test Set'], loc='upper right')
ax[0][1].set_ylim(-0.2, 7.0)
ax[0][1].plot(np.full((10,), best_epoch), np.linspace(-0.5, 7.0, 10), c='black')
ax[0][1].set_xticks([best_epoch, 20, 40, 60, 80, 100])
ax[0][1].set_xticklabels(['Optimal', '20', '40', '60', '80', '100'])

ax[1][0].plot(CNN_real2real_pC_acc[0:12])
ax[1][0].plot(CNN_real2real_pC_val_acc[0:12])
ax[1][0].set_title('CNN Accuracy - Optimal Training Period')
ax[1][0].set_ylabel('Accuracy')
ax[1][0].set_xlabel('Epoch')
ax[1][0].legend(['Training Set', 'Test Set'], loc='lower right')
ax[1][0].set_ylim(-0.05, 1.05)
ax[1][0].plot(np.full((10,), best_epoch), np.linspace(-0.5, 1.5, 10), c='black')
ax[1][0].set_xticks([0, 2, 4, 6, 8, 10])
ax[1][0].set_xticklabels(['0', '2', '4', 'Optimal', '8', '10'])


ax[1][1].plot(CNN_real2real_pC_loss[0:12])
ax[1][1].plot(CNN_real2real_pC_val_loss[0:12])
ax[1][1].set_title('CNN Cost - Optimal Training Period')
ax[1][1].set_ylabel('Cost')
ax[1][1].set_xlabel('Epoch')
ax[1][1].legend(['Training Set', 'Test Set'], loc='upper right')
ax[1][1].set_ylim(-0.2, 7.0)
ax[1][1].plot(np.full((10,), best_epoch), np.linspace(-0.5, 7.0, 10), c='black')
ax[1][1].set_xticks([0, 2, 4, 6, 8, 10])
ax[1][1].set_xticklabels(['0', '2', '4', 'Optimal', '8', '10'])

#plt.savefig('../plots/results/thesis/NN-overfitting.pdf')
```
